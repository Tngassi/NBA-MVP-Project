```python
# Some players were on multiple teams in a specific season. For example, in 2016 D.J. Augustin played OKC and DEN. The TOT label is the total for that season. Here I want to consolidate the rows
# into a single line for the season and attach the latest team he played for that season. 

def single_row(df):
    if df.shape[0] == 1:
        return  df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
# Adding in team nicknames from the nicknames.csv file 
nicknames = {}
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

Here am I looking at how different statistics correlate to winning MVP. I'm excluding "PTS Won", "PTS Max", "Share" as they direct contributors to winning MVP since the MVP is based on votes. Then taking those correlations values and plotting the values. 

```python
corr = {}
for column in stats.columns:
    if stats[column].dtype == "float64":
        corr[column] = stats["Share"].corr(stats[column])

print(corr)
```
